```python
contacts = {
    "Jenish": "0412 345 678",
    "Chandu": "0423 567 890",
    "Shishir": "0434 678 901"
}

contacts["Ram"] = "0455 789 012"

print("Contact List:")
for name in sorted(contacts):
    print(name + ":", contacts[name])
```

```python
contacts = {
    "Jenish": {"phone": "0412 345 678", "email": "jenish@example.com"},
    "Chandu": {"phone": "0423 567 890", "email": "chandu@example.com"},
    "Shishir": {"phone": "0434 678 901", "email": "shishir@example.com"}
}

contacts["Ram"] = {"phone": "0455 789 012", "email": "ram@example.com"}

print("Contact Details:\n")

for name, info in contacts.items():
    print(name)
    print(" Phone:", info["phone"])
    print(" Email:", info["email"])
    print()
```

I used AI to understand how to store multiple details using a nested dictionary. I modified the code by updating the names, phone numbers, and emails to match my own data.
