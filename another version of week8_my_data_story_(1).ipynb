# Week 8 – My Data Story

This notebook reads a text file called `my_data.txt`, displays its contents, and calculates simple statistics.


## AI collaboration note

I used AI to help me structure the notebook, write the Python code for reading the file, and calculate statistics such as the number of lines, words, and characters. I reviewed the code and adjusted the text content so it matched my own daily routine examples.


```python
# Create the text file so the notebook works in Colab
content = [
    "Wake up early and go for a walk",
    "Prepare healthy breakfast at home",
    "Attend university lectures on time",
    "Complete programming practice tasks",
    "Work on assignment submissions",
    "Take short breaks to stay focused",
    "Go to part-time work shift",
    "Learn new skills online",
    "Spend time with friends or family",
    "Plan tasks for the next day"
]

with open("my_data.txt", "w", encoding="utf-8") as file:
    for line in content:
        file.write(line + "\n")

print("my_data.txt created successfully.")
```

```python
# Read and display the file
filename = "my_data.txt"

with open(filename, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("Contents of my_data.txt:")
for line in lines:
    print(line.strip())
```

```python
# Calculate statistics
number_of_lines = len(lines)
number_of_words = sum(len(line.split()) for line in lines)
number_of_characters = sum(len(line.strip()) for line in lines)

print("\nStatistics:")
print("Number of lines:", number_of_lines)
print("Number of words:", number_of_words)
print("Number of characters:", number_of_characters)
```
